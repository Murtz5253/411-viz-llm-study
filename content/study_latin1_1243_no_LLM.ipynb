A small Seattle-based coffee shop chain has five locations. The operations manager wants to share a year-end review with the regional team summarizing how each location performed in 2024. They asked an analyst on their team to make a chart showing each location's monthly transactions, with circle size encoding the typical customer rating so the team can quickly see whether busy locations also keep customers happy.

The dataset includes a customer_rating column with values from 1 to 5, representing the modal customer rating for that location-month, where 1 = Very Dissatisfied, 2 = Dissatisfied, 3 = Neutral, 4 = Satisfied, and 5 = Very Satisfied.

The analyst produced the script below. The chart runs without throwing errors, but several issues—both syntactic (a coding/labeling mistake) and conceptual (an error rooted in a misunderstanding of how visualizations should be designed)—have made it into the final visualization. You can see that the generated visualization, while rendered, does not make sense.


### Your Task
Copy the code below into the provided blank code cell after the following cell. **Do not edit the cell below.** Once you have done so. indentify and fix all syntactic and conceptual errors you can find in the code. For each error, **identify** it in the Markdown cell after as well as **explain how you corrected it.** Note that you should identify errors even if you do not know hot to fix them, as you can receive partial credit.

```python
coffee_df = pd.read_csv('coffee_sales.csv')


month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']


chart = alt.Chart(coffee_df).mark_circle().encode(
    x=alt.X(
        'month:O',
        sort=month_order,
        title='Month',
        scale=alt.Scale(domain=[0, 20])
    ),
    y=alt.Y(
        'transactions:Q',
        title='Transactions',
        axis=alt.Axis(labelPadding=60)
    ),
    size=alt.Size(
        'mean(customer_rating):Q',
        title='Avg Customer Rating',
        scale=alt.Scale(range=[50, 400])
    ),
    color=alt.Color(
        'location:N',
        scale=alt.Scale(scheme='blues'),
        legend=None
    ),
    tooltip=['location', 'month', 'transactions', 'avg_ticket_usd', 'customer_rating']
).properties(
    width=600,
    height=400,
    title='Monthly Transactions by Location, 2024'
)


chart
```
